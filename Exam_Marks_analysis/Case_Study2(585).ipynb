**IMPORTING LIBRARIES**


```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

**Loading DataSet --> MIDMARKS (excel file)**

```python
df = pd.read_excel('MIDMARKS.xlsx')
df
```

**Filling NaN with (0's)**

```python
df.fillna(0, inplace=True)
df
```

**'S.NO' type changed to integer**

```python
df["S.NO"]=df["S.NO"].astype("int64")
df["S.NO"]=range(1,len(df)+1)
df
```

**Replacing All AB, MP, A as '0'**

```python
df = df.replace("A", 0)
df = df.replace("AB", 0)
df = df.replace("MP", 0)

df = df.infer_objects(copy=False) 
df
```

**Adding a new Column 'Total' for each individual student**

```python
df["Total"]=df["DV"]+df["MA"]+df["PP"]+df["BEEE"]+df["FL"]+df["FIMS"]
df
```

**Obtained By 120 marks actual total**

```python
df['Obtained By'] = 120
df
```

**Student's having atleast one '0'**

```python
subject_columns = ['DV', 'MA', 'PP', 'BEEE', 'FL', 'FIMS']
df['Check_if_0'] = (df[subject_columns] == 0).any(axis = 1)
df
```

**Backlog Count**

```python
subject_columns = ['DV', 'MA', 'PP', 'BEEE', 'FL', 'FIMS']  
df['Backlog'] = (df[subject_columns] < 8).sum(axis=1)
df
```

**Failed in all**

```python
df['Failed_All'] = (df[subject_columns] < 8).all(axis=1)
df
```

**-----------------------------------------------------------------------------------------**

**All Subject scores are combined together as a column Total**

```python
def assign_grade(Total):
    if Total >= 110:
        return 'A+'
    elif Total >= 100:
        return 'A'
    elif Total >= 90:
        return 'B'
    elif Total >= 80:
        return 'C'
    elif Total >= 70:
        return 'D'
    elif 50 <= Total <= 60: 
        return 'E'
    else:
        return 'Fail'

df['Grade'] = df['Total'].apply(assign_grade)
df
```

**Grades are assigned to the students total scores**

```python
df['SECTION'] = df['SECTION'].fillna('SIGMA')  
df['SECTION'] = df['SECTION'].replace('', 'SIGMA')
df['SECTION'] = df['SECTION'].replace(0, 'SIGMA') 
df
```
