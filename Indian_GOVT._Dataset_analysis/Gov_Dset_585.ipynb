**IMPORTING LIBRARIES**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Using Pandas

**Importing the dataset**


```python
df = pd.read_csv("Indian_Air_Traffic_Status(1).csv")
df
```

**Understandng shape of dataframe**

```python
df.shape
```

**Understand Daataframe by using .info() function**

```python
df.info()
```

**Total No. of countries**

```python
x = df['Country'].value_counts().sum()
print("Total countries: ",x)
```

**Operation by Indian carriers is null**

```python
n = (df["Indian_Carriers"] == 0).sum()
print("Number of countries: ",n)
```

**Operation by Foreign carriers is null**

```python
m = (df["Foreign_Carriers"] == 0).sum()
print("Number of countries: ",m)
```

**Country with Maximum Departures & Arrivals**

```python
max_departure = df.loc[df["Indian_Carriers"].idxmax(), ["Country", "Indian_Carriers"]]
max_arrival = df.loc[df["Foreign_Carriers"].idxmax(), ["Country", "Foreign_Carriers"]]
print(f"Maximum Departures: {max_departure['Country']} with {max_departure['Indian_Carriers']} seats per week")
print(f"Maximum Arrivals: {max_arrival['Country']} with {max_arrival['Foreign_Carriers']} seats per week")
```

**Finding the Country with Maximum Departures (Indian Carriers)**

```python
max_departure = df.loc[df["Indian_Carriers"].idxmax(), ["Country", "Indian_Carriers"]]
print(f"Top Country for Departures: {max_departure['Country']} with {max_departure['Indian_Carriers']} seats/week")
```

**Find the Country with Maximum Arrivals (Foreign Carriers)**

```python
max_arrival = df.loc[df["Foreign_Carriers"].idxmax(), ["Country", "Foreign_Carriers"]]
print(f"Top Country for Arrivals: {max_arrival['Country']} with {max_arrival['Foreign_Carriers']} seats/week")
```

**Country with Least Departures**

```python
min_departure = df.loc[df["Indian_Carriers"].idxmin(), ["Country", "Indian_Carriers"]]
print(f"Least Departures: {min_departure['Country']} with {min_departure['Indian_Carriers']} seats per week")
```

**Country with Least Arrivals**

```python
min_arrival = df.loc[df["Foreign_Carriers"].idxmin(), ["Country", "Foreign_Carriers"]]
print(f"Least Arrivals: {min_arrival['Country']} with {min_arrival['Foreign_Carriers']} seats per week")
```

**Country with least Departures > 0**

```python
min_departure = df[df["Indian_Carriers"] > 0].loc[df[df["Indian_Carriers"] > 0]["Indian_Carriers"].idxmin(), ["Country", "Indian_Carriers"]]
print(f"Least Departures: {min_departure['Country']} with {min_departure['Indian_Carriers']} seats per week")
```

**Country with least Arrivals > 0**

```python
min_arrival = df[df["Foreign_Carriers"] > 0].loc[df[df["Foreign_Carriers"] > 0]["Foreign_Carriers"].idxmin(), ["Country", "Foreign_Carriers"]]
print(f"Least Arrivals: {min_arrival['Country']} with {min_arrival['Foreign_Carriers']} seats per week")
```

**Top 5 Countries by Indian Carriers’ Seats**

```python
top_5_indian = df.nlargest(5, "Indian_Carriers")[["Country", "Indian_Carriers"]]
print("Top 5 Countries by Indian Carriers' Seats:\n", top_5_indian)
```

**Top 5 Countries by Foreign Carriers’ Seats**

```python
top_5_foreign = df.nlargest(5, "Foreign_Carriers")[["Country", "Foreign_Carriers"]]
print("Top 5 Countries by Foreign Carriers' Seats:\n", top_5_foreign)
```

**Average Seats by Indian Carriers**

```python
avg_indian_seats = df["Indian_Carriers"].mean()
print(f"Average Seats Operated by Indian Carriers per Country: {avg_indian_seats:.2f}")
```

**Average Seats by Foreign Carriers**

```python
avg_foreign_seats = df["Foreign_Carriers"].mean()
print(f"Average Seats Operated by Foreign Carriers per Country: {avg_foreign_seats:.2f}")
```

# Using Matplotlib

**Boxplot (Distribution of Seats for Indian & Foreign Carriers)**

```python
plt.figure(figsize=(10,8))
plt.boxplot([df["Indian_Carriers"], df["Foreign_Carriers"]], tick_labels=["Indian Carriers", "Foreign Carriers"])
plt.title("Boxplot of Seat Distribution")
plt.ylabel("Seats per Week")
plt.grid(True)
plt.show()
```

**Scatter Plot (Indian vs. Total Seats comparision)**

```python
df["Total_Seats"] = df["Indian_Carriers"] + df["Foreign_Carriers"]

plt.figure(figsize=(7, 5))
plt.scatter(df["Indian_Carriers"], df["Total_Seats"], color="green", alpha=0.7, edgecolors="black")

plt.xlabel("Seats by Indian Carriers")
plt.ylabel("Total Seats Available (Indian + Foreign)")
plt.title("Scatter Plot: Indian Carriers vs Total Seats")
plt.grid(True)
plt.show()
```

**Bar Graph (Seats per Country - Indian Carriers)**

```python
filtered_df = df[(df["Indian_Carriers"] > 0) | (df["Foreign_Carriers"] > 0)]

plt.figure(figsize=(12, 8))
plt.bar(filtered_df["Country"], filtered_df["Indian_Carriers"], color="orange")
plt.xlabel("Country")
plt.ylabel("Seats per Week")
plt.title("Seats Operated by Indian Carriers per Country")
plt.xticks(rotation=45, ha="right")
plt.show()
```

**Bar Graph (Seats per Country - Foreign Carriers)**

```python
plt.figure(figsize=(12,8))
plt.bar(df["Country"], df["Foreign_Carriers"], color="blue")
plt.xlabel("Country")
plt.ylabel("Seats per Week")
plt.title("Seats Operated by Foreign Carriers per Country")
plt.xticks(rotation=45, ha="right")
plt.show()
```

**Line Plot (Indian vs Foreign Carriers' Seats Trend)**

```python
plt.figure(figsize=(13,8))
plt.plot(df["Country"], df["Indian_Carriers"], marker="o", linestyle="-", color="blue", label="Indian Carriers")
plt.plot(df["Country"], df["Foreign_Carriers"], marker="s", linestyle="--", color="orange", label="Foreign Carriers")
plt.xlabel("Country")
plt.ylabel("Seats per Week")
plt.title("Line Plot: Indian vs Foreign Carriers' Seats")
plt.xticks(rotation=45, ha="right")
plt.legend()
plt.grid(True)
plt.show()
```

**Histogram (Grouping Countries by Seat Availability - Indian Carriers**

```python
plt.figure(figsize=(6,4))
plt.hist(df["Indian_Carriers"], bins=[0, 1000, 3000, 5000, 7000], color="purple", alpha=0.7, edgecolor="black")
plt.xlabel("Seats per Week (Indian Carriers)")
plt.ylabel("Number of Countries")
plt.title("Distribution of Indian Carriers' Seats")
plt.xticks([500, 2000, 4000, 6000], labels=["0-1K", "1K-3K", "3K-5K", "5K-7K"])
plt.grid(axis="y")
plt.show()
```

**Histogram (Grouping Countries by Seat Availability - Foreign Carriers**

# Using Seaborn

** Pie Chart (Indian Carriers' Seat Share by Country)**

```python
plt.figure(figsize=(6,6))
df_sorted = df.sort_values(by="Indian_Carriers", ascending=False).head(7)  # Top 7 for better visualization
plt.pie(df_sorted["Indian_Carriers"], labels=df_sorted["Country"], autopct="%1.1f%%", colors=sns.color_palette("pastel"))
plt.title("Indian Carriers' Seat Share by Country")
plt.show()
```

**Pie Chart (Foreign Carriers' Seat Share by Country)**

**Heatmap (Indian vs Foreign Carriers Seat Correlation)**

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[["Indian_Carriers", "Foreign_Carriers"]].corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap: Indian vs Foreign Carriers' Seats")
plt.show()
```

**Seaborn Bar Plot (Top 5 Countries with Highest Indian Seats)**

```python
df_filtered = df[~df["Country"].str.contains("Total", case=False, na=False)]

top_indian = df_filtered.nlargest(5, "Indian_Carriers")

plt.figure(figsize=(8, 5))
sns.barplot(x="Country", y="Indian_Carriers", data=top_indian, hue="Country", palette="Blues_r", legend=False)
plt.xlabel("Country")
plt.ylabel("Seats per Week")
plt.title("Top 5 Countries with Highest Indian Carriers' Seats (Excluding Total)")
plt.xticks(rotation=45, ha="right")
plt.show()
```

**Pie Chart - Distribution of Indian Carriers' Seat Ranges**

```python
bins = [0, 2000, 4000, 6000, 8000, float('inf')]
labels = ["0-2K", "2K-4K", "4K-6K", "6K-8K", "8K+"]

df["Indian_Seat_Range"] = pd.cut(df["Indian_Carriers"], bins=bins, labels=labels, right=False)
seat_distribution = df["Indian_Seat_Range"].value_counts()

plt.figure(figsize=(7,7))
plt.pie(seat_distribution, labels=seat_distribution.index, autopct="%1.1f%%", colors=sns.color_palette("Set2"), startangle=140)
plt.title("Distribution of Indian Carriers' Seat Ranges")
plt.show()
```

**Pie Chart - Distribution of Foreign Carriers' Seat Ranges**

```python
bins = [0, 2000, 4000, 6000, 8000, float('inf')]
labels = ["0-2K", "2K-4K", "4K-6K", "6K-8K", "8K+"]

df["Foreign_Seat_Range"] = pd.cut(df["Foreign_Carriers"], bins=bins, labels=labels, right=False)
seat_distribution = df["Foreign_Seat_Range"].value_counts()

plt.figure(figsize=(7,7))
plt.pie(seat_distribution, labels=seat_distribution.index, autopct="%1.1f%%", colors=sns.color_palette("Set2"), startangle=140)
plt.title("Distribution of Foreign Carriers' Seat Ranges")
plt.show()
```
